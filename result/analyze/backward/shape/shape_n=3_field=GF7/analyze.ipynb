```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df_grevlex = pd.read_csv('test.grevlex.csv', index_col=0)
df_grlex = pd.read_csv('test.grlex.csv', index_col=0)
df_lex = pd.read_csv('test.lex.csv', index_col=0)
```

```python
df_grevlex['reduction_rate'] = df_grevlex['reduction_time'] / df_grevlex['total_time']
df_grlex['reduction_rate'] = df_grlex['reduction_time'] / df_grlex['total_time']
df_lex['reduction_rate'] = df_lex['reduction_time'] / df_lex['total_time']

df_grevlex['selection_rate'] = df_grevlex['selection_time'] / df_grevlex['total_time']
df_grlex['selection_rate'] = df_grlex['selection_time'] / df_grlex['total_time']
df_lex['selection_rate'] = df_lex['selection_time'] / df_lex['total_time']
```

```python
# Plot the relationship between polynomial_additions and total_time for each dataframe
plt.figure(figsize=(12, 6))
column_name = 'polynomial_additions'
y_label_name = 'Polynomial Additions'

plt.subplot(1, 3, 1)
# plt.scatter(df_grevlex['polynomial_additions'], df_grevlex['total_time'], alpha=0.5)
plt.scatter(df_grevlex['total_time'], df_grevlex[column_name], alpha=0.5)

plt.title('Grevlex')
plt.xlabel('Total Time')
plt.ylabel(y_label_name)


plt.subplot(1, 3, 2)
# plt.scatter(df_grlex['polynomial_additions'], df_grlex['total_time'], alpha=0.5)
plt.scatter(df_grlex['total_time'], df_grlex[column_name], alpha=0.5)

plt.title('Grlex')
plt.xlabel('Total Time')
plt.ylabel(y_label_name)


plt.subplot(1, 3, 3)
plt.scatter(df_lex['total_time'], df_lex[column_name], alpha=0.5)
plt.title('Lex')
plt.xlabel('Total Time')
plt.ylabel(y_label_name)


plt.tight_layout()
plt.show()
```
