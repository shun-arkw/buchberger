```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_grevlex = pd.read_csv('test.grevlex.csv', index_col=0)
df_grlex = pd.read_csv('test.grlex.csv', index_col=0)
# df_lex = pd.read_csv('test.lex.csv', index_col=0)
```
